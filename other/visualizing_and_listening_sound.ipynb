# Wokring Code for audio visualization

```python
import numpy as np
import random
import itertools
import librosa
import IPython.display as ipd
import matplotlib.pyplot as plt



def load_audio_file(file_path):
#     input_length = 16000   # Sample_rate 
    input_length = 22050
    data = librosa.core.load(file_path)[0] #, sr=16000
    if len(data)>input_length:
        data = data[:input_length]
    else:
        data = np.pad(data, (0, max(0, input_length - len(data))), "constant")
    return data


def plot_time_series(data):
    fig = plt.figure(figsize=(14, 8))
    plt.title('Raw wave ')
    plt.ylabel('Amplitude')
    plt.plot(np.linspace(0, 1, len(data)), data)
    plt.show()
    
```

```python
data = load_audio_file("./Breathing-Sounds/wheeze/wheeze (36).wav")
plot_time_series(data)
```

```python
file = "./Breathing-Sounds/wheeze/wheeze (36).wav"
# file = "intro.mp4"
sr = 22050         # by default sample rate for analyzing audio data.

signal, sr = librosa.load(file, sr)   # signal_values = sample_rate * length_of_sound
print(len(signal))
# signal is going to be numpy one dimesional array. 
# sr * T => 22050 * 15 = 3000k 
# so we r gonna get 3000k values, and value mean amplituide points on waveform

librosa.display.waveplot(signal, sr = sr)
plt.xlabel("Time")
plt.ylabel("Amplituide")
plt.show()
```
